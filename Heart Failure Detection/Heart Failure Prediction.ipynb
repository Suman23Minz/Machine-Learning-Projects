```python
# DATA PREPROCESSING...
```

```python
dataset = pd.get_dummies(df, columns=['Sex','ChestPainType','FastingBS','RestingECG','ExerciseAngina','ST_Slope'])
```

```python
# FEATURE SCALING...
```

```python
# Retriving the data from database.
dataset.columns
```

```python
# Importing StandardScaler object from sklearn library.
from sklearn.preprocessing import StandardScaler
stndSclr = StandardScaler()

# Identifying column to scale.
column_to_scale = ['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']

# Standardising the selected column, fit_transform method is applied.
dataset[column_to_scale] = stndSclr.fit_transform(dataset[column_to_scale])
```

```python
# Returns the first x number of rows using head(x), default is 5.
dataset.head()
```

```python
# Splitting the dataset into dependent and independent features.
X = dataset.drop('HeartDisease', axis = 1)
y = dataset['HeartDisease']
```

```python
# MODEL BUILDING
```

```python
# KNeighbors Classifier Model...
```

```python
# Importing KNeighborsClassifier from sklearn library.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Initializing an empty list.
forest_scores = []

# Finding the best accuracy for knn algorithm using cross_val_score
for i in range(10, 101, 10):
    forest_classifier = RandomForestClassifier(n_estimators=i)
    cvs_scores = cross_val_score(forest_classifier, X, y, cv=5)
    forest_scores.append(round(cvs_scores.mean(),3))
```

```python
# Plotting the results of knn_scores
plt.figure(figsize=(20,15))
plt.plot([k for k in range(10, 101, 10)], forest_scores, color = 'red')
for i in range(1,11):
    plt.text(i*10, forest_scores[i-1], (i*10, forest_scores[i-1]))
plt.xticks([i for i in range(10, 101, 10)])
plt.xlabel('Number of Estimators (N)')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different N values')
```

```python
# Training the random forest classifier model with n value as 90
forest_classifier = RandomForestClassifier(n_estimators=90)
cvs_scores = cross_val_score(forest_classifier, X, y, cv=5)
print("Random Forest Classifier Accuracy with n_estimators=90 is: {}%".format(round(cvs_scores.mean(), 4)*100))
```
